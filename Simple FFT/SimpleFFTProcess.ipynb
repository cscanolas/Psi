```python
import numpy as np
import matplotlib.pyplot as plt
```

Preliminary Information and information from the book

```python
# Preliminary information from the book

path = 'Data/3 unmodelled.txt'
data = np.loadtxt(path, dtype=float, delimiter=",")

duration = 1200
samplingFreq = 100
samplingPeriodT = 1/samplingFreq
Nsamples = duration * samplingFreq

#extract the parameters
sdof_1 = data[:,0]
sdof_2 = data[:,1]
sdof_3 = data[:,2]

print("{}...{}".format(sdof_1[0:3], sdof_1[-3:]))
print("{}...{}".format(sdof_2[0:3], sdof_2[-3:]))
print("{}...{}".format(sdof_3[0:3], sdof_3[-3:]))
```

```python
timeVector = np.arange(0, duration, samplingPeriodT)

fig, [ax1, ax2, ax3] = plt.subplots(nrows=3, ncols=1)

ax1.plot(timeVector, sdof_1)
ax2.plot(timeVector, sdof_2)
ax3.plot(timeVector, sdof_3)

plt.show()
```

FFT

```python
sdof_1FFT = np.fft.fft(sdof_1)
sdof_2FFT = np.fft.fft(sdof_2)
sdof_3FFT = np.fft.fft(sdof_3)

freqBinVector = np.arange(0, samplingFreq, 1/duration)

fig, [ax1, ax2, ax3] = plt.subplots(nrows=3, ncols=1)

ax1.plot(freqBinVector, sdof_1FFT)
ax2.plot(freqBinVector, sdof_2FFT)
ax3.plot(freqBinVector, sdof_3FFT)

plt.show()
```
