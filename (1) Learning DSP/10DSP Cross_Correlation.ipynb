# **10 DSP Cross Correlation**

## **[1] Preliminary**
Import modules

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal
```

### **[1.1] Create Signals**
`np.r_` - concatenates from start : last : step in terms of addition
*   [<u>np.r_ documentation </u>](https://numpy.org/doc/stable/reference/generated/numpy.r_.html)

```python
print("Array from -pi to pi (not including pi) with pi/8 incrementation:\n{}".format(np.r_[ -(np.pi) : (np.pi) : (np.pi/8) ]))

print("\nNormalized array:\n{}".format(np.r_[ -(np.pi) : (np.pi) : (np.pi/8) ] / (np.pi)))
```

```python
#create first signal
sig1 = np.sin(np.r_[ -(np.pi) : (np.pi) : (np.pi/8) ] )

# Second signal with pi/4 phase shift.
# Half the size of sig1
sig2 = np.sin(np.r_[ -(np.pi) : 0 : (np.pi/8) ] + np.pi/4 )
```

### **[1.2] Plot the signal**

```python
plt.figure(1)
plt.plot(sig1,'bo-', label = "sig1")
plt.plot(sig2,'r*-', label = "sig2")
plt.legend()
plt.grid()
plt.show()
```

## **[2] Cross correlation**

### **[2.1] Cross correlation - numpy implementation**

* `np.correlate(a, v, mode='valid)` - cross correlation of two 1-dimensional sequences  
  * a and v are sequences being zero padded
  * [ <u>np.correlate documentation </u>](https://numpy.org/doc/stable/reference/generated/numpy.correlate.html)



Defined by the equation: 
$$ c_k = \sum_n a_{n+k} \bar v_n $$



```python
# Method #1  Numpy implementation

corr1 = np.correlate( a=sig1, v=sig2 )
print( corr1 )
print( "Method#1:position of maximum in ", corr1.argmax(axis=0))
```

### **[2.2] Cross correlation - scipy implementation**

* `scipy.signal.correlate( in1, in2, mode='full', method='auto' )`
  * cross correlate two N-dimensional arrays
  * Cross-correlate in1 and in2, with the output size determined by the mode argument.
  * [<u>scipy documentation link</u>](https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.correlate.html)



```python
corr2 = scipy.signal.correlate( sig1, sig2 )
print( corr2 )
print( "Method#2:position of maximum in ", corr2.argmax(axis=0))
```
